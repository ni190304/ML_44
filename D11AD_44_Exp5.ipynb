```python
glass_types = ['Building Windows (float processed)', 
               'Building Windows (non-float processed)', 
               'Vehicle Windows (float processed)', 
               'Vehicle Windows (non-float processed)', 
               'Containers', 
               'Tableware']
glass_counts = [df['Glass_type'].value_counts().get(1, 0),df['Glass_type'].value_counts().get(2, 0),df['Glass_type'].value_counts().get(3, 0),df['Glass_type'].value_counts().get(4, 0),df['Glass_type'].value_counts().get(5, 0),df['Glass_type'].value_counts().get(6, 0), ]

plt.figure(figsize=(8, 6))
plt.pie(glass_counts, labels=glass_types, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Glass Types')
plt.axis('equal')
plt.show()
```

```python
import seaborn as sns

sns.pairplot(df, hue='Glass_type', diag_kind='hist', palette='husl')
plt.show()
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Features')
plt.show()
```

```python
X = df.drop('Glass_type', axis=1)
y = df['Glass_type']
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix

non_linear_model = SVC(kernel='rbf', random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.2)

non_linear_model.fit(X_train, y_train)
```

```python
np.array(y_test)
```

```python
y_preds = non_linear_model.predict(X_test)
y_preds
```

```python
cm = confusion_matrix(y_test, y_preds)

plt.figure(figsize=(4, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="YlOrBr", cbar=False, 
            xticklabels=['Predicted 0', 'Predicted 1'], 
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
non_linear_model.score(X_train, y_train)
```

```python
non_linear_model.score(X_test, y_test)
```

```python
np.random.seed(42)

cv_acc = cross_val_score(non_linear_model, X, y, cv=10, scoring="accuracy")
cv_acc
```
